```python
from pyit2fls import zero_mf, singleton_mf, const_mf, tri_mf, ltri_mf, rtri_mf, trapezoid_mf, gaussian_mf
import matplotlib.pyplot as plt
from numpy import linspace

domain_ringan = linspace(0., 5, 1000)
domain_sedang = linspace(4, 10, 1000)
domain_berat = linspace(8, 14, 1000)


trapezoid = trapezoid_mf(domain_sedang, [4, 6, 8, 10, 1.0])
ltri = tri_mf(domain_ringan, [0, 4, 5, 1.0])
rtri = tri_mf(domain_berat, [8, 10, 14, 1.0 ])

plt.figure()
plt.plot(domain_ringan, ltri, label="Ringan")
plt.plot(domain_sedang, trapezoid, label="Sedang")
plt.plot(domain_berat, rtri, label="Berat")
plt.grid(True)
plt.legend()
plt.xlabel("Berat Pakaian")
plt.ylabel("Membership function")
plt.savefig("M04_5200411486_Adie Gunawan Alwani_Berat Pakaian.png")
plt.show()
```
